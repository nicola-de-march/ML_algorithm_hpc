```python
import pandas as pd
import matplotlib.pyplot as plt

#read from csv
file1 = 'results_serial_Forecaster.csv'
file2 = 'results_128MPI_Forecaster.csv'

data1 = pd.read_csv(file1) #serial
data2 = pd.read_csv(file2) #mpi version

# Filter data to ensure numerical values
data1 = data1[pd.to_numeric(data1['Number_of_Forecasters'], errors='coerce').notna()] #if an entry is not-numeric, we put Nan
data2 = data2[pd.to_numeric(data2['Number_of_Forecasters'], errors='coerce').notna()] #keep only real number (not NaN)
data1 = data1[pd.to_numeric(data1['Execution Time (s)'], errors='coerce').notna()]  #same
data2 = data2[pd.to_numeric(data2['Execution Time (s)'], errors='coerce').notna()]
data1['Number_of_Forecasters'] = data1['Number_of_Forecasters'].astype(int) #convert to int
data1['Execution Time (s)'] = data1['Execution Time (s)'].astype(float) #convert to float
data2['Number_of_Forecasters'] = data2['Number_of_Forecasters'].astype(int)
data2['Execution Time (s)'] = data2['Execution Time (s)'].astype(float)

# extraction of right columns
forecasters_1 = data1['Number_of_Forecasters']
execution_time_1 = data1['Execution Time (s)']
forecasters_2 = data2['Number_of_Forecasters']
execution_time_2 = data2['Execution Time (s)']

# we create the chart
plt.figure(figsize=(10, 6))

#line for serial
plt.plot(forecasters_1, execution_time_1, marker='o', linestyle='-', label='Serial')

#line for mpi
plt.plot(forecasters_2, execution_time_2, marker='s', linestyle='--', label='128 MPI')

# chart information
plt.xlabel('Number of Forecasters')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time vs Number of Forecasters')
plt.legend()

# grid
plt.grid(True, linestyle='--', alpha=0.7)
#show
plt.show()
```
